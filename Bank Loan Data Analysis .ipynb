 ### DATA CLEANING 

```python
# Fill numeric columns with median
numeric_columns = df.select_dtypes(include='number').columns.tolist()
for col in numeric_columns:
    df[col].fillna(df[col].median(), inplace=True)
 
```

```python
# Fill categorical columns with mode
cat_columns = df.select_dtypes(include='object').columns.tolist()
for col in cat_columns:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
# Fix date columns
date_cols = ['issue_date', 'last_credit_pull_date',
             'last_payment_date', 'next_payment_date']
for col in date_cols:
    df[col] = pd.to_datetime(df[col], format='%d-%m-%Y', errors='coerce')
 
```

```python
 # Fix int_rate — remove % sign if present
if df['int_rate'].dtype == object:
    df['int_rate'] = df['int_rate'].str.replace('%', '').astype(float)
```

```python
# Fix term — extract numeric value
if df['term'].dtype == object:
    df['term'] = df['term'].str.extract(r'(\d+)').astype(int)
```

```python
# Fix emp_length — map to numeric
emp_length_map = {
    '< 1 year': 0, '1 year': 1, '2 years': 2, '3 years': 3,
    '4 years': 4,  '5 years': 5, '6 years': 6, '7 years': 7,
    '8 years': 8,  '9 years': 9, '10+ years': 10
}
```

```python
df['emp_length'] = df['emp_length'].map(emp_length_map)
df['emp_length'].fillna(df['emp_length'].median(), inplace=True)
 
```

 
#### ── FEATURE ENGINEERING ───

```python
# Loan Category — Good Loan vs Bad Loan
df['loan_category'] = df['loan_status'].apply(
    lambda x: 'Good Loan' if x in ['Fully Paid', 'Current'] else 'Bad Loan'
)
print(df['loan_category'].unique())
 
```

```python
# Date-based features
df['issue_month']      = df['issue_date'].dt.month          # Numeric month (1–12)
df['issue_year']       = df['issue_date'].dt.year           # Numeric year
df['issue_month_name'] = df['issue_date'].dt.month_name()   # Month name
df['issue_weekday']    = df['issue_date'].dt.day_name()     # Day name
 
```

```python
# Loan size category
def loan_size(amount):
    if amount <= 5000:
        return 'Small'
    elif amount <= 15000:
        return 'Medium'
    else:
        return 'Large'
 
df['loan_size'] = df['loan_amount'].apply(loan_size)
df['loan_size']
```

```python
# Income category
def income_category(income):
    if income <= 40000:
        return 'Low'
    elif income <= 80000:
        return 'Medium'
    else:
        return 'High'
 
df['income_category'] = df['annual_income'].apply(income_category)
df['income_category']
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Numeric version for plotting
Monthly_Funded_amount = round(df.groupby('issue_month_name').agg(
    Funded_amount=('loan_amount', 'sum'),
    Amount_received=('total_payment', 'sum'),
    Profit_gain=("net_cash_flow", "sum")
) / 1_000_000, 2).sort_values(by="Funded_amount")

# Sort by calendar month order
month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']
Monthly_Funded_amount.index = pd.CategoricalIndex(
    Monthly_Funded_amount.index, categories=month_order, ordered=True
)
Monthly_Funded_amount = Monthly_Funded_amount.sort_index()

# Convert index to datetime for time series
Monthly_Funded_amount.index = pd.to_datetime(Monthly_Funded_amount.index, format='%B')

# Plot
fig, axes = plt.subplots(3, 1, figsize=(10, 8), sharex=True)

# --- Funded Amount ---
axes[0].plot(Monthly_Funded_amount.index, Monthly_Funded_amount['Funded_amount'],
             marker='o', linewidth=2, color='steelblue')
axes[0].fill_between(Monthly_Funded_amount.index, Monthly_Funded_amount['Funded_amount'],
                     alpha=0.2, color='steelblue')
for x, y in zip(Monthly_Funded_amount.index, Monthly_Funded_amount['Funded_amount']):
    axes[0].annotate(f'{y:.2f}M', xy=(x, y), ha='center', va='bottom', fontsize=8, color='steelblue')
axes[0].set_title('Funded Amount Over Time')
axes[0].set_ylabel('Amount (M)')
axes[0].grid(axis='y', linestyle='--', alpha=0.5)
axes[0].yaxis.set_major_formatter(plt.FuncFormatter(lambda v, _: f'{v:.0f}M'))

# --- Amount Received ---
axes[1].plot(Monthly_Funded_amount.index, Monthly_Funded_amount['Amount_received'],
             marker='s', linewidth=2, color='coral')
axes[1].fill_between(Monthly_Funded_amount.index, Monthly_Funded_amount['Amount_received'],
                     alpha=0.2, color='coral')
for x, y in zip(Monthly_Funded_amount.index, Monthly_Funded_amount['Amount_received']):
    axes[1].annotate(f'{y:.2f}M', xy=(x, y), ha='center', va='bottom', fontsize=8, color='coral')
axes[1].set_title('Amount Received Over Time')
axes[1].set_ylabel('Amount (M)')
axes[1].grid(axis='y', linestyle='--', alpha=0.5)
axes[1].yaxis.set_major_formatter(plt.FuncFormatter(lambda v, _: f'{v:.0f}M'))

# --- Profit Gain ---
axes[2].plot(Monthly_Funded_amount.index, Monthly_Funded_amount['Profit_gain'],
             marker='^', linewidth=2, color='mediumseagreen')
axes[2].fill_between(Monthly_Funded_amount.index, Monthly_Funded_amount['Profit_gain'],
                     alpha=0.2, color='mediumseagreen')
for x, y in zip(Monthly_Funded_amount.index, Monthly_Funded_amount['Profit_gain']):
    axes[2].annotate(f'{y:.2f}M', xy=(x, y), ha='center', va='bottom', fontsize=8, color='mediumseagreen')
axes[2].set_title('Profit Gain Over Time')
axes[2].set_ylabel('Amount (M)')
axes[2].grid(axis='y', linestyle='--', alpha=0.5)
axes[2].yaxis.set_major_formatter(plt.FuncFormatter(lambda v, _: f'{v:.0f}M'))

# Format x-axis as month names
axes[2].xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%B'))
plt.xticks(Monthly_Funded_amount.index, rotation=45, ha='right')

fig.suptitle('Time Series — Monthly Financial Overview', fontsize=14, fontweight='bold', y=1.01)
plt.tight_layout()
plt.show()
```

```python
monthly_summary = df.groupby("issue_month_name").agg(
    id           = ("id", "count"),
    loan_amount  = ("loan_amount", "sum"),
    total_payment= ("total_payment", "sum"),
    int_rate     = ("int_rate", "mean"),
    dti          = ("dti", "mean")
).reset_index().sort_values(by="loan_amount")

# Add Profit column before formatting
monthly_summary["profit"] = round((monthly_summary["total_payment"] - monthly_summary["loan_amount"]) / 1000000, 2)

# Format columns
monthly_summary["loan_amount"]   = round(monthly_summary["loan_amount"]   / 1000000, 2)
monthly_summary["total_payment"] = round(monthly_summary["total_payment"] / 1000000, 2)
monthly_summary["int_rate"]      = round(monthly_summary["int_rate"]      * 100, 2)
monthly_summary["dti"]           = round(monthly_summary["dti"]           * 100, 2)

# Add M and % signs
monthly_summary["loan_amount"]   = monthly_summary["loan_amount"].astype(str)   + " M"
monthly_summary["total_payment"] = monthly_summary["total_payment"].astype(str) + " M"
monthly_summary["profit"]        = monthly_summary["profit"].astype(str)        + " M"
monthly_summary["int_rate"]      = monthly_summary["int_rate"].astype(str)      + " %"
monthly_summary["dti"]           = monthly_summary["dti"].astype(str)           + " %"

print("monthly_summary")
print(monthly_summary)
```
